ЗАДАЧА 1 Б
Из колоды в 52 карты извлекаются случайным образом 4 карты. б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.

ОПИСАНИЕ ОШИБКИ: не верное решение, для решения задачи пойдем от обратного и вычтем из всех возможных комбинаций, комбинации без тузов

```python
import numpy as np
from math import factorial

def combinations(n, k):
    return int(factorial(n)/(factorial(k)*factorial(n-k)))

C = combinations(52, 4) - combinations(48, 4)
P = C/combinations(52, 4)
f"Вероятность, что среди 4-х карт окажется хотя бы один туз P = {P}"
```

В результате 10 независимых измерений некоторой величины X, выполненных с одинаковой точностью, получены опытные данные:
6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1
Предполагая, что результаты измерений подчинены нормальному закону распределения вероятностей,оценить истинное значение
величины X при помощи доверительного интервала, покрывающего это значение с доверительной вероятностью 0,95.

ОШИБКА: неверный расчет, ход решения правильный (решал устно, ошибка в расчетах)

```python
X = np.array([6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1])

#сигма не известно, используем критерий Стьюдента

std = np.std(X,ddof=1)
t = 2.262
inter = t*std/(len(X)**0.5)

[np.mean(X) - inter,np.mean(X) + inter]
```
